```python
from urllib.request import urlretrieve
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
#import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
# initialize list of lists
data = [['hard','2019_1','1','1','2','tom', 'jim'], ['hard','2019_1','2','3','4','bill', 'jeff'],['clay','2019_2','3','1','2','tom', 'jim'], ['clay','2019_2','4','4','3','jeff', 'bill'],['hard','2019_3','5','1','2','tom', 'jim'], ['hard','2019_3','6','4','3','jeff', 'bill'],['hard','2019_4','7','1','5','tom', 'chris'],['hard','2019_5','8','1','2','tom', 'jim'],['hard','2019_6','9','2','1','jim', 'tom'],['hard','2019_7','10','2','1','jim', 'tom']]
 
# Create the pandas DataFrame
df = pd.DataFrame(data, columns = ['Surface', 'Tour_Week','match_id','w_ID','l_ID','Winner', 'Loser'])
df
```

```python
df_winners = df
df_winners["Outcome"] = 1
df_winners = df_winners.rename(columns={'Winner': 'Player','Loser': 'Opponent','w_ID':'player_ID','l_ID':'opponent_ID'})
df_winners
```

```python
df_losers = df
df_losers["Outcome"] = 0
df_losers = df_losers.rename(columns={'Loser': 'Player','Winner': 'Opponent','l_ID':'player_ID','w_ID':'opponent_ID'})
df_losers
```

```python
df_players = pd.concat([df_winners, df_losers], ignore_index=True)
df_players
```

```python
df_players = df_players.sort_values(by=['Tour_Week'], ascending = False)
df_players
```

```python
df_players = df_players.iloc[::-1]
df_players['past_W'] = df_players.groupby(['player_ID','opponent_ID','Surface'])['Outcome'].transform(lambda x: x.rolling(window=100, min_periods = 1).sum().shift(1))
df_players = df_players.iloc[::-1]
df_players['past_W'] = df_players['past_W'].fillna(0)
df_players
```

```python
# Split back into winner and loser dataframes transiently
df_winners2 = df_players[df_players['Outcome'] == 1]
df_losers2 = df_players[df_players['Outcome'] == 0]
```

```python
# Now we can just do a left join on m_id
df_players2 = df_winners2.merge(df_losers2, on='match_id', how = 'left')
df_players2
```
