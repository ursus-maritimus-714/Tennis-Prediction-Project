```python
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
#import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load preprocessed
df = pd.read_csv('../data/df_player_4.csv')
```

```python
df.info()
```

```python
df2 = df[['t_name','t_id','p_sv_pts_won%']]
df2
```

```python
df2 = df2.sort_values(by=['t_id'], ascending = False)
df2
```

```python
df3 = df2.groupby(['t_id','t_name']).mean().round(2)
```

```python
df3.head(20)
```

```python
df3 = df3.sort_values(by=['t_name','t_id'], ascending = False)
df3
```

```python
df3["t_sv_pts_won%_last"] = df3.groupby('t_name')['p_sv_pts_won%'].shift(-1)
df3
```

```python
#Save current df
df3.to_csv('../data/test_df3.csv', index=True)
```
