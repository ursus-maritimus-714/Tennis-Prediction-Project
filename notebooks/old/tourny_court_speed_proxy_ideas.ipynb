```python
import pandas as pd
import numpy as np
#import matplotlib.pyplot as plt
#import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load preprocessed
df = pd.read_csv('../data/df_player_4.csv')
```

```python
df.info()
```

```python
df2 = df[['t_name','t_id','p_ace%']]
df2
```

```python
df2 = df2.sort_values(by=['t_id'], ascending = False)
df2
```

```python
df3 = df2.groupby(['t_id','t_name']).mean().round(2)
```

```python
df3.head(20)
```

```python
df3 = df3.sort_values(by=['t_name','t_id'], ascending = False)
df3
```

```python
df3.rename(columns = {'p_ace%':'t_ace%'}, inplace=True)
```

```python
df3["t_ace%_last"] = df3.groupby('t_name')['t_ace%'].shift(-1)
df3
```

```python
# Now we can just do a left join on m_id
df4 = df.merge(df3['t_ace%_last'], on='t_id', how = 'left')
```

```python
#Save current df
df4.to_csv('../data/test_df4.csv', index=True)
```
